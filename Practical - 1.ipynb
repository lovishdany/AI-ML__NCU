```python
import pandas as pd

data = pd.read_csv("https://raw.githubusercontent.com/cmdlinetips/data/master/us_states.tsv", sep="\t")
df = data.head()

dict1 = dict(zip(df.state, df.latitude))
print(dict1)
dict2 = df.set_index('state').to_dict()['name']
print("\n",dict2,"\n")
```
