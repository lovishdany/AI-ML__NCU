```python
import numpy as np
import pandas as pd

data = pd.read_csv('/content/Data.csv')
df = pd.DataFrame(data)
df1 = df.copy()
df2 = df.copy()
df
```

```python
print(df.isna())
```

```python
print("Total no. of Nan values : ",df.isnull().sum().sum())
```

```python
col_list = list(df.columns)

null_cols = []
for i in col_list:
  if df[i].isnull().values.any() == True:
    null_cols.append(i)
null_cols
```

```python
from sklearn.impute import SimpleImputer

imp = SimpleImputer(strategy="mean")
for col in null_cols:
  df[col] = imp.fit_transform(df[col].values.reshape(-1,1))
df
```

```python
df['Age'] = df['Age'].astype(int)
df['Salary'] = df['Salary'].round(decimals = 2)
df
```

```python
imp = SimpleImputer(strategy="most_frequent")
for col in null_cols:
  df1[col] = imp.fit_transform(df1[col].values.reshape(-1,1))
df1
```

```python
imp = SimpleImputer(strategy="median")
for col in null_cols:
  df2[col] = imp.fit_transform(df2[col].values.reshape(-1,1))
df2
```
