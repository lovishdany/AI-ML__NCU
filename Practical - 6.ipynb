```python
X1 = Advertising["TV"]
X2 = Advertising["Radio"]
X3 = Advertising["Newspaper"]

y = Advertising["Sales"]
x_list = X1.to_numpy()
y_list = y.to_numpy()
x_list.reshape(-1,1)
y_list.reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y, test_size = 0.2, random_state=42)
X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y, test_size = 0.2, random_state=12)
X3_train, X3_test, y3_train, y3_test = train_test_split(X3, y, test_size = 0.2, random_state=54)
```

```python
from matplotlib import pyplot as plt
def plot_predictions(X_train, y_train, X_test, y_test, y_preds):
  plt.scatter(X_train, y_train, c="b", label="Training")
  plt.scatter(X_test, y_test, c="g", label="Testing")
  plt.plot(X_test, y_preds, c="r", label="Predictions")
  plt.legend()
  plt.show()
```

```python
#Linear Regression using Tensorflow
import tensorflow as tf
def Advertising(X_train, y_train, X_test, y_test):
  Advertising_model = tf.keras.Sequential([
      tf.keras.layers.Dense(1, activation = "linear")
  ])

  Advertising_model.compile(
      loss = tf.keras.losses.mae,
      optimizer = tf.keras.optimizers.Adam(learning_rate=0.01)
  )

  Advertising_model.fit(X_train,y_train,epochs=100, verbose = 0)
  preds = Advertising_model.predict(X_test)
  plot_predictions(X_train, y_train, X_test, y_test, preds)
Advertising(X1_train, y1_train, X1_test, y1_test)
```
