```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "X":[78,80,81,82,84,80],
    "Y":[126,128,127,130,130,132]
})

X_mean = df["X"].mean()
Y_mean = df["Y"].mean()

standard_df = df.copy()
standard_df["X"] = standard_df["X"] - X_mean
standard_df["Y"] = standard_df["Y"] - Y_mean
standard_df
```

```python
cov = standard_df.cov()
cov_arr = cov.to_numpy()
cov_arr
```

```python
eig_value, eig_vector = np.linalg.eig(cov_arr)
eig_value, eig_vector
```
